```python
import json


factor_file_name = "encoded_2016-06-01-2017-06-01all-factors.json"
with open(factor_file_name, "r") as fin:
    factor_data = json.load(fin)
    
sector_file_name = "encoded_2016-06-01-2017-06-01all-sectors.json"
with open(sector_file_name, "r") as fin:
    sector_data = json.load(fin)
```

```python
def save_json(ids, coods, name):
 
    json_dict = dict()
    for i in range(ids.shape[0]):
        json_dict[int(ids[i])] = {"x": float(coods[i,0]), "y": float(coods[i, 1])}
        
    with open(name, "w", encoding='utf-8') as f:
        json.dump(json_dict, f)
```

```python
import numpy as np
ids = np.array(list(factor_data.keys()))
X = np.array(list(factor_data.values()))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline  



from sklearn import decomposition
plt.figure(1, figsize=(20, 20))
# X is the PCA result
pca = decomposition.PCA(n_components=2)
pca.fit(X)
low_X = pca.transform(X)
plt.plot(low_X[:, 0], low_X[:, 1], "b+")

plt.show()

save_json(ids, low_X, "PCA" + factor_file_name)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline  
from sklearn import decomposition
from sklearn.manifold import TSNE


plt.figure(1, figsize=(20, 20))
# X is the PCA result
model = TSNE(n_components=2)
low_X = model.fit_transform(X)

plt.plot(low_X[:, 0], low_X[:, 1], "b+")

plt.show()

save_json(ids, low_X, "TSNE" +  factor_file_name)
```
